# functools

It is a tool which helps to work with higher order functions (functions that take another function as input or return them)


```python
#Faster & more memory efficient than class
#class is mutable and namedtuple is immutable
class User:
    def __init__(self,name,age,country):
        self.name=name
        self.age=agefrom functools import lru_cache

#lru_cache is adecorator. It remebers the result if functions
#maxsize=3 denotes it will remember last 3 unique numbers
# for eg: fibonnaci series using recursive fucntion will recalculated a numbers multiple times
#fib(6) calls fib(5) & fib(4), fib(5) calls fib(4) & fib(3); fib(4) is calculated again
#lru-least recently used

@lru_cache(maxsize=3)
def fib(n):
    if n<=1:
        return n
    return fib(n-1)+fib(n-2)
print(fib(6))

#lru_Cache use memoization
#memoization is a technique where you store result of expensive function calls
#Reuse them when the same input occurs again


#Creates a new function with some argument fixed
from functools import partial

def power(base,exp):
    return base**exp

square=partial(power,exp=2)
print(square(5))

#square=partial(power,exp) is equivalent to 
# def square(base,exp):
#   return power(base,exponent=2)
#but without rewriting the function


from functools import reduce

#reduce applies to function cumulatively to items in a sequence
#Syntax: reduce(function,iterable)
nums=[1,2,3,4]
result=reduce(lambda x,y:x+y,nums)
print(result)


from functools import wraps

# wraps is used inside decorators to preserve the original function's information like: function name, docstring, annotations


        self.country=country
```
