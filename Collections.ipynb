# Collections

#### Counter
Counter is a dictionary subclass used to count hasable objects

```python
#Count of element 
arr=[1,2,3,3,4,4]
counts={}
for x in arr:
    counts[x]=counts.get(x,0)+1
counts
```

```python
#The above line of code is similar to the following
#It gices an output in dictionary form, where key represent the element and the value represent the occurrence
#In Counter missing key will return 0 value
from collections import Counter

#Counter can take in list
print(Counter([1,2,3,3,4,4]))

#Counter can take in dictionary
print(Counter({1:3,2:4}))

#Input string/keyword arguments as well
print(Counter('asbdajs'))
```

```python
#Common methods in counter

#most_coomon returns common elements sorted by frequency
print(Counter([1,2,2.1,1,1,1]).most_common(1))

#updating counts
c=Counter([1,2,3])
print(c)
c.update([4,5,5])
print(c)

#deleting/removing elements
del c[2]

#ARTHEMTIC OPERATIONS
#Addition
print(Counter(a=3,b=1) + Counter(a=1,b=2))

#Subtraction
print(Counter(a=3,b=1)-Counter(a=1,b=2))

#Intersection &-min counts
print(Counter(a=3,b=1)&Counter(a=1,b=2))

#Union-max counts
print(Counter(a=1,b=4)|Counter(a=4,b=6))
```

#### nametuple()

It is alightweight immutable data structure that lets you access tuple elements by name instead of index. In other words, we can say that it is a tuple with named fields

```python
#Problem with normal tuple
user=('Sonali',28,'India')
user[0] #What is this? name? city? Age?
#hard to read, error-prone, bad for large codebase

#Solution - namedtuple()
#Readable, clean, self-documenting
from collections import namedtuple
User=namedtuple('User',['name','age','country'])
user=User('Sonali',28,'India')
print(user.name)
print(user.age)
print(user.country) 

print('\n')
print(user[0])
print(user.name)
```

```python
#Properties
#Immutable
user.age=30
```
